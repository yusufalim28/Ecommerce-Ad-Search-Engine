```python
print("\n" + "="*70)
print("DEMO: SISTEM PENCARIAN SEDERHANA")
print("="*70)

#  dokumen dari dataset
documents = [
"aku mau tau lebih lanjut Benefit Scentplus Previlage Card ini dong, bantu aku join yaaa, sebelum slot-nya habis 🖤",

"Aku tertarik mau order Black Katalog dengan harga spesial di WA Official Scentplus. Boleh dibantu untuk proses ordernya sebelum kehabisan?",

"Mau claim voucher Ramadhan Special Offer Warm Greatness 100ML hanya 220.000. Bisa dibantu proses   klaim voucher WhatsApp-nya yaa?",

"Join Scentplus Club Sekarang → Hadiah Langsung Aktif!",

"🔥 Mau harga paling hemat",

"💳 Info Privilege Card & benefitnya.",

"✨ Bantu aku pilih varian Black Catalog yang sesuai denganku",

"🤔 Bantu aku pilih varian Black Catalog yang cocok",

"💥 Mau Special Deal Cuma Buat Aku",

"Aku baru dapet Voucher Diskon Eksklusif Black Katalog, Boleh dibantuin buat redeem sekarang? sebelum kuotanya habis🥹"
]
print("\nDaftar Dokumen:")
print("─"*70)
for i, doc in enumerate(documents, 1):
    print(f"\nDocument {i}:")
    print(f"   {doc}")

print("\n" + "="*70)
print("Jumlah dokumen:", len(documents))
print("*dokumen di ambil dari landing page dan tautan iklan otomatis di whatsapp comerce salah satu brand di indonesia.")
print("""
   Proses yang akan dilakukan:
   1. Preprocessing teks
   2. Representasi Bag of Words (BoW)
   3. Menghitung kemiripan dokumen dengan Cosine Similarity
   4. Membuat Search / Similarity Engine
""")

print("="*70)
```

```python
import re
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
factory = StemmerFactory()
stemmer = factory.create_stemmer()


factory = StemmerFactory()
stemmer = factory.create_stemmer()

processed_docs = []
for i, doc in enumerate(documents, 1):

    print(f"\nDocument {i}")

    # Original
    print("Original   :", doc)

    # Lowercase
    lower = doc.lower()
    print("Lowercase  :", lower)

    # Cleaning
    cleaning = re.sub(r'[^a-zA-Z\s]', '', lower)
    print("Cleaning   :", cleaning)

    # Tokenizing
    tokens = cleaning.split()
    print("Tokenizing :", tokens)

    # Stemming
    stemmed = [stemmer.stem(word) for word in tokens]
    print("Stemming   :", stemmed)

    # Final text
    final = " ".join(stemmed)
    print("Final Text :", final)

    processed_docs.append(final)

    print(f"\nDocument {i}")
    print("Original :", doc)
    print("Processed:", final)
```

```python
print("\n" + "="*70)
print("MEMBANGUN VOCABULARY")
print("="*70)

# kumpulkan semua kata
vocab_set = set()

for doc in processed_docs:
    tokens = doc.split()
    vocab_set.update(tokens)

# ubah ke list dan urutkan
vocabulary = sorted(list(vocab_set))

print("\nVocabulary yang terbentuk:")
print("─"*70)

for i, word in enumerate(vocabulary):
    print(f"{i+1}. {word}")

print("\nJumlah kata unik:", len(vocabulary))

print("="*70)
```

## **BoW ENGINEER**

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print("\n" + "="*70)
print("IMPLEMENTASI COUNT VECTORIZER")
print("="*70)

# Membuat objek CountVectorizer
# memindai seluruh processed_docs, mencari kata unik, dan menyiapkan kolom untuk setiap kata tersebut.
vectorizer = CountVectorizer()

# Mengubah dokumen menjadi vektor angka
X = vectorizer.fit_transform(processed_docs)

# Mengambil daftar kata
terms = vectorizer.get_feature_names_out()

# Membuat Document Term Matrix
dtm = pd.DataFrame(
    X.toarray(),
    columns=terms,
    index=[f"Doc_{i+1}" for i in range(len(processed_docs))]
)

plt.figure(figsize=(10, 6))
sns.heatmap(dtm, annot=True, cmap='YlOrRd', fmt='d', cbar_kws={'label': 'Frequency'})

print("\nDocument Term Matrix:")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
print("─"*70)
print(dtm)
```

```python
print("\n" + "="*70)
print("KATA PALING SERING MUNCUL")
print("="*70)

word_freq = dtm.sum().sort_values(ascending=False)

plt.figure(figsize=(10,5))

word_freq.head(10).plot(kind='bar')

plt.title("Top 10 Kata Paling Sering Muncul")
plt.xlabel("Kata")
plt.ylabel("Frekuensi")

plt.xticks(rotation=45)

plt.show()
```

 **SIMILARITY ENGINEER**

```python
print("\n" + "="*70)
print("SIMILARITY ENGINE - COSINE SIMILARITY ANTAR DOKUMEN")
print("="*70)
from sklearn.metrics.pairwise import cosine_similarity

# Menghitung cosine similarity dari Document-Term Matrix
similarity_matrix = cosine_similarity(X)

# Membuat tabel agar lebih mudah dibaca
similarity_df = pd.DataFrame(
    similarity_matrix,
    index=[f"Doc{i+1}" for i in range(len(processed_docs))],
    columns=[f"Doc{i+1}" for i in range(len(processed_docs))]
)

print("\nTabel Kemiripan Antar Dokumen:\n")
print(similarity_df.round(3))

# Nilai Similarity
# 1.000	Dokumen sama dengan dirinya sendiri
# 0.5 – 0.9	Sangat mirip
# 0.2 – 0.4	Cukup mirip
# 0 – 0.1	Hampir tidak mirip
```

```python
print("\n" + "="*70)
print("DOKUMEN PALING MIRIP")
print("="*70)

doc_index = 1   # misalnya kita ingin melihat dokumen yang mirip dengan Doc1

similar_scores = list(enumerate(similarity_matrix[doc_index]))

# urutkan dari similarity tertinggi
similar_scores = sorted(similar_scores, key=lambda x: x[1], reverse=True)

print(f"\nDokumen yang paling mirip dengan Doc{doc_index+1}:\n")

for i, score in similar_scores:  # top 3 selain dirinya sendiri
    print(f"Doc{i+1} dengan similarity = {score:.3f}")
```

```python
print("\n" + "="*70)
print("VISUALISASI SIMILARITY MATRIX")
print("="*70)

import matplotlib.pyplot as plt

# Membuat figure
plt.figure(figsize=(8,6))

# Menampilkan similarity matrix sebagai heatmap
plt.imshow(similarity_matrix)

# Menambahkan colorbar
plt.colorbar()

# Label sumbu
plt.xticks(range(len(processed_docs)), [f"Doc{i+1}" for i in range(len(processed_docs))])
plt.yticks(range(len(processed_docs)), [f"Doc{i+1}" for i in range(len(processed_docs))])

# Judul
plt.title("Heatmap Kemiripan Antar Dokumen (Cosine Similarity)")

plt.show()
```

**SEARCH** **ENGINE**

```python
print("\n" + "="*70)
print("SEARCH ENGINE DEMO")
print("="*70)

from sklearn.metrics.pairwise import cosine_similarity

queries = [
    "voucher diskon",
    "black catalog",
    "privilege card benefit"
]

for q in queries:

    print("\n" + "-"*60)
    print(f"Query : {q}")
    print("-"*60)

    # ubah query menjadi vector
    query_vector = vectorizer.transform([q])

    # hitung similarity query dengan semua dokumen
    similarity_scores = cosine_similarity(query_vector, X)

    # ubah ke list + urutkan
    scores = list(enumerate(similarity_scores[0]))
    scores = sorted(scores, key=lambda x: x[1], reverse=True)

    print("\nTop 3 Dokumen Paling Relevan:\n")

    for i, score in scores[:3]:
        print(f"Doc{i+1} | Similarity = {score:.3f}")
        print(f"Isi dokumen : {processed_docs[i]}\n")
```
